```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
import pandas as pd
```

```python
val=pd.read_csv("iris.csv")
x=val.iloc[:,:4].values
print(x)
y=val.iloc[:,4].values
print(y)
```

```python
print(val.describe)
```

```python
print(val.shape)
print(val.head(5))
print(val.tail(2))
```

```python
le=LabelEncoder()
ynew=le.fit_transform(y)
print(ynew)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,ynew,test_size=0.25,random_state=2)
print(xtrain)
print("\n")
print(ytrain)
print("\n")
print(xtest)
print("\n")
print(ytest)
print("\n")
```

```python
mlp=MLPClassifier(hidden_layer_sizes=(30,30),activation="relu",max_iter=1000)
mlp.fit(xtrain,ytrain)
ypredict=mlp.predict(xtest)
print(ytrain)
print(ypredict)
```

```python
print(accuracy_score(ytest,ypredict))
```
