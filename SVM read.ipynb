```python
from sklearn.svm  import SVC
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
import pandas as pd
```

```python
val=pd.read_csv("iris.csv")
x=val.iloc[:,:4].values
print(x)
y=val.iloc[:,4].values
print(y)
```

```python
le=LabelEncoder()
ynew=le.fit_transform(y)
print(ynew)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,ynew,test_size=0.25,random_state=2)
print(xtrain)
print("\n")
print(ytrain)
print("\n")
print(xtest)
print("\n")
print(ytest)
print("\n")
```

```python
svm=SVC(kernel="linear",random_state=2)
svm.fit(xtrain,ytrain)
ypredict=svm.predict(xtest)
print(ytest)
print(ypredict)
```

```python
print(accuracy_score(ytest,ypredict))
```
