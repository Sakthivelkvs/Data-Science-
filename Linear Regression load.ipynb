```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
var=load_iris()
xval=var.data[:,0]
y=var.target
print(xval,"\n")

x=xval.reshape(-1,1)
print(x)

print(y)
```

```python
print("Frature Name-------:\n",var.feature_names,"\n")
print("Target Name--------:\n",var.target_names)
```

```python
print(var.DESCR)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=1)
print(xtrain)
print(ytrain)
```

```python
li_reg=LinearRegression()
li_reg.fit(xtrain,ytrain)
ypredict=li_reg.predict(xtest)
print(ytest)
print(ypredict)
```

```python
print(li_reg.coef_)
print(li_reg.intercept_)
```

```python
plt.scatter(xtest,ytest,marker=".",color="green")
plt.plot(xtest,ypredict,marker="*",color="red")
plt.xlabel("Acual Value")
plt.ylabel("Predicted Value")
```

```python
print(mean_squared_error(ytest,ypredict))
print(r2_score(ytest,ypredict))
```
