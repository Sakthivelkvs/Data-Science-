```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("iris.csv")
x=data.iloc[:,:4].values
print(x)
y=data.iloc[:,4].values
print(y)
```

```python
le=LabelEncoder()
y_new=le.fit_transform(y)
print(y_new)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y_new,test_size=0.25,random_state=2)
print(xtrain)
print("\n")
print(ytrain)
print("\n")
print(xtest)
print("\n")
print(ytest)
print("\n")
```

```python
knn=KNeighborsClassifier(n_neighbors=12)
```

```python
knn.fit(xtrain,ytrain)
```

```python
ypredict=knn.predict(xtest)
print(ytest)
print(ypredict)
```

```python
print(accuracy_score(ytest,ypredict,normalize=True))
```

```python
cm=confusion_matrix(ytest,ypredict,labels=knn.classes_)
print(cm)
```

```python
cmd=ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=knn.classes_)
cmd.plot()
plt.show()
```
